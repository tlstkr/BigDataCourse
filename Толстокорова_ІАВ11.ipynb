Набір даних зпарсений з сайту "https://www.formula1.com/en/drivers.html" за допомогою пакету Beautiful Soup:

```python
url = "https://www.formula1.com/en/drivers.html"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content)
table = soup.find('table')
```

```python
driver_links = []
for i in soup.select('a[class="driver"]'):
    driver_links.append(i.get('href').split('/')[-1])
```

```python
formula1_list = []
for i in driver_links:
    url_d = 'https://www.formula1.com/en/drivers/' + i
    response_d = requests.get(url_d)
    soup_d = BeautifulSoup(response_d.content)
    table_d = soup_d.find('table')
    biography = []
    info_d = []
    for i in soup_d.select('div[class="text parbase"] p'):
        biography.append(i.text)    
    for i in soup_d.select('td[class="stat-value"]'):
        info_d.append(i.text)
    formula1_list.append([" ".join(biography)] + info_d)
```

```python
formula1 = pd.DataFrame(formula1_list, columns =['Biography', 'Team', 'Country', 'Podiums', 'Points', 'Grands Prix entered', 
                                                 'World Championships', 'Highest race finish', 'Highest grid position', 
                                                 'Date of birth', 'Place of birth'])
formula1
```
